```python
import pandas as pd
```

```python
df = pd.read_csv("/home/desktop/Downloads/DWFormer-main/download/result.xlsx.csv")
df.head()
```

```python
valid = df[df['type']=='valid']
valid.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

# define data values
# display
```

```python
for i in range(1,6):
    s = valid[valid['session']==i]
    x = s['epoch']  # X-axis points
    y = s['wa']  # Y-axis points
    z = s['ua'] # second
    #plt.title("Session id:",i)
    plt.plot(x, y,color='r',label='wa')
    plt.plot(x,z,color='g',label='ua')  # Plot the chart
    plt.xlabel("epoches")
    plt.ylabel("Accuracy Score in %")
    plt.title("Session id")
    plt.show()  
    
```

```python
## main focus for the paper is on the valid score and we are arriving on the close
```

```python
valid['best'] = valid['ua'] + valid['wa']
valid.head()
```

```python
best_wa=[]
best_ua=[]

for i in range(1,6):
    s = valid[valid['session']==i]
    maxValueIndex = s['best'].idxmax()
    best_ua.append(s['ua'][maxValueIndex])
    best_wa.append(s['wa'][maxValueIndex])


```
